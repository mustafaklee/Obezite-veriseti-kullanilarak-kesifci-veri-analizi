```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#obez olan kadın ve erkeklerin yaş gruplarınca incelenmesi 
```

```python
young_ob=df[(df.Age>17) & (df.Age<26) & (df.ObesityCategory=="Obese")]

middle_ob=df[(df.Age>=26) & (df.Age<50) & (df.ObesityCategory=="Obese")]

old_ob=df[(df.Age>=50) & (df.ObesityCategory=="Obese")]

g1=sns.lmplot(x = "Age", y = "BMI",hue="Gender", data = young_ob);
g2=sns.lmplot(x = "Age", y = "BMI",hue="Gender", data = middle_ob);
g3=sns.lmplot(x = "Age", y = "BMI",hue="Gender", data = old_ob);
g1.fig.suptitle('Obez olan kadın ve erkeklerin yaş gruplarınca incelenmesi (Age:18 ve 25) ', fontsize=12, y=1.02);
g2.fig.suptitle('Obez olan kadın ve erkeklerin yaş gruplarınca incelenmesi (Age:26 ve 49) ', fontsize=12, y=1.02);
g3.fig.suptitle('Obez olan kadın ve erkeklerin yaş gruplarınca incelenmesi (Age:50+) ', fontsize=12, y=1.02);
```

```python
#Normal kiloda olan genç kadın ve erkeklerin fiziksel aktivitelerinin incelenmesi
```

```python
normal_young=df[(df.ObesityCategory=="Normal weight") & (df.Age>17) & (df.Age<25)]
```

```python
sns.barplot(x = "Gender", y = df.PhysicalActivityLevel, data= normal_young);
plt.title("Normal kiloda olan genç kadın ve erkeklerin fiziksel aktivitelerinin incelenmesi");
```

```python
#Zayıf kiloda olan genç kadın ve erkeklerin fiziksel aktivitelerinin incelenmesi
```

```python
weak_young=df[(df.ObesityCategory=="Underweight") & (df.Age>17) & (df.Age<25)]
```

```python
sns.barplot(x = "Gender", y = df.PhysicalActivityLevel, data= weak_young);
plt.title("Zayıf kiloda olan genç kadın ve erkeklerin fiziksel aktivitelerinin incelenmesi");
```

```python
#normal kiloda olan yaşlı kişilerin bilgileri
```

```python
obese_old=df[(df.ObesityCategory=="Obese") & (df.Age>64)]
```

```python
sns.barplot(x = "Gender", y = df.PhysicalActivityLevel, data= obese_old);
plt.title("Obez olan yaşlı insanların fiziksel aktivitelerinin incelenmesi");
```
